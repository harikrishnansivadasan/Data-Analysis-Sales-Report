# EXPLORATORY DATA ANALYSIS

## Gender

```python
df.columns
```

```python
colors = ["pink", "skyblue"]
ax = sns.countplot(x = 'Gender',data = df, hue='Gender',palette=colors )
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Gender'] ,as_index= False)
```

```python
sales_gen_sum = sales_gen['Amount_spend'].sum().sort_values(['Amount_spend'], ascending=False)
```

```python
print(sales_gen_sum)
```

```python
spendchart=sns.barplot(x='Gender',y='Amount_spend' , data=sales_gen_sum ,hue='Gender',palette=colors)
for bars in spendchart.containers:
    spendchart.bar_label(bars)
```

From above graph it is shown that more purchase has been done by Females


## AGE

```python
df.columns
```

```python
ax= sns.countplot(x='Age Group', data = df , hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount_spend'].sum().sort_values(by='Amount_spend',ascending=False)
sales_age.describe
```

```python
sales_age.info()
```

```python
ax=sns.barplot(x='Age Group',y='Amount_spend', data=sales_age, hue='Age Group')
for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs we can see that most of the buyers are of age group between 26-35

```python
# Filter the DataFrame for the age group 26-35
age_26_35 = df[df['Age Group'] == '26-35']
age_26_35.head()
age_26_35.info()
```

```python
# Group by 'Gender' and sum 'Amount_spend'
sales_gen_age_group = age_26_35.groupby('Gender')['Amount_spend'].sum().reset_index()
```

```python
ax=sns.barplot(x='Gender',y='Amount_spend',data=sales_gen_age_group, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)

# Set labels and title
plt.xlabel('Age Group 26-35')
plt.ylabel('Total Amount Spent')
```

In Age group 26-35 total most of the buyers are females

## State vs Total Spend


```python
# total number of orders from top 10 states
df.columns
```

```python
state_sales=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
```

```python
state_sales.describe
```

```python
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data=state_sales,  x='State', y='Orders',hue='State')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
state_sales=df.groupby(['State'], as_index=False)['Amount_spend'].sum().sort_values(by='Amount_spend', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data=state_sales,  x='State', y='Amount_spend',hue='State')
for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs most of the sales and orders are from uttar pradesh,maharashtra and karnataka respectively

## Marital status vs Amount Spend

```python
ax = sns.countplot(x='Marital_Status', data=df, hue='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

## Occupation vs Amount spend


```python
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(x='Occupation', data=df, hue='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Number of coustomers from each Occupations')
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount_spend'].sum().sort_values(by='Amount_spend', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount_spend',hue='Occupation')
plt.title('Amount_spend vs Occupation')
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

## Product Category

```python
df.info()
```

```python
df.tail(5)
```

```python
ax=sns.countplot(x='Product_Category', data=df, hue='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(35,15)})
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount_spend'].sum().sort_values(by='Amount_spend', ascending=False)
sns.set(rc={'figure.figsize':(35,15)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount_spend', hue='Product_Category')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders', hue='Orders')
```
